```r
x = seq(0, 2*pi, length.out=100)
y = sin(x)
df = data.frame(x,y)

ggplot(df, aes(x,y)) +
    geom_point() +
    theme_bw() + 
    labs(
        title= "y = sin(x)",
        x = "X axis",
        y = "Y axis",
        caption = "explain here",
        tag ="A",
        color = "Curve"
    ) +

    theme(
        plot.title = element_text(size = rel(2), face="bold"),
        axis.title = element_text(size=rel(2)),
        axis.text = element_text(size=rel(1.25)),
        legend.title = element_text(size=rel(1.5)),
        legend.text = element_text(size=rel(1.5)),
        legend.position = "right"
    )
```

```r
# 必要なライブラリをロードします
library(ggplot2)
library(dplyr)
library(tidyr)

# データを作成します
df <- data.frame(x = seq(-pi, pi, length.out = 100))

# サインカーブとコサインカーブを追加します
df <- df %>%
  mutate(sin = sin(x),
         cos = cos(x))

# データフレームを整形して、ggplotで扱いやすい形にします
df_melt <- df %>%
  gather(key = "curve", value = "value", -x)

# グラフを描きます
ggplot(df_melt, aes(x = x, y = value, color = curve)) +
  geom_line() +
  scale_color_manual(values = c("sin" = "blue", "cos" = "red")) +
  labs(x = "X", y = "Y",
       color = "Curve",
       title = "Sin and Cos Curves") +
  theme(
    legend.position = "right",
    legend.title = element_text(size=rel(1.25)),
    legend.text = element_text(size=rel(1.25))
  )
```
