```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as pat

a = 41.54463
b = 16.96493
```

```python
b
```

```python
fig = plt.figure(figsize=(6, 6))

ax = fig.add_subplot(111)
ax.set_xlim(60, 80)
ax.set_ylim(150, 190)


E1 = pat.Ellipse(xy=(71.45, 164.7), width=a/2, height=b/2, color="lime", alpha=0.5, angle=0.3*180)

ax.add_patch(E1)
plt.grid()
plt.show()
```
