```python
import numpy as np
from scipy.spatial.distance import cdist

# 
n = 100 # 
dim = 2 # 
range_param = 0.5 # 
smoothness_param = 1.5 # 

# 
X = np.random.uniform(size=(n,dim))

# 
D = cdist(X, X)

# 
C = np.exp(-range_param*D) * (1 + range_param*D) ** (-smoothness_param)

# 
if dim == 2:
    import matplotlib.pyplot as plt
    plt.scatter(X[:,0], X[:,1], c=C.flatten())
    plt.colorbar()
    plt.show()
```
